```python
import requests
from bs4 import BeautifulSoup

def get_selector_from_url(url, selector):
    req = requests.get(url)
    assert req.status_code == 200, f"Failed to fetch {url}"
    soup = BeautifulSoup(req.text, 'html.parser')
    element = soup.select(selector)
    assert len(element) == 1, f"Selector {selector} did not match exactly one element in {url}"
    return element[0].get_text()
```

```python
import hashlib

def get_hash_from_url(text):
    return hashlib.md5(text.encode()).hexdigest()

def get_selectors_from_dict(url_dict, url):
    if not url in url_dict:
        raise ValueError(f"URL {url} not found in the provided dictionary.")
    item = url_dict[url]

    req = requests.get(url)
    assert req.status_code == 200, f"Failed to fetch {url}"
    soup = BeautifulSoup(req.text, 'html.parser')

    result = {
        "url": url,
        "paragraphs": []
    }

    for selector in item:
        element = soup.select(selector)
        assert len(element) == 1, f"Selector {selector} did not match exactly one element in {url}"
        hash = get_hash_from_url(element[0].get_text())
        result["paragraphs"].append({
            "text": element[0].get_text(),
            "selector": selector,
            "hash": hash
        })
    return result
```

```python
import json

pages = []
for page in MALLACHTAI:
    pages.append(get_selectors_from_dict(MALLACHTAI, page))
with open("mallachta.json", "w") as f:
    f.write(json.dumps(pages, indent=4))
```
