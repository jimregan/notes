```python
import srt
```

```python
SRT_DIR = "/tmp/srt"
```

```python
def srt_to_json(srtfile):
    segments = []
    with open(srtfile) as sf:
        for sub in srt.parse(sf.read()):
            segments.append({
                "id": sub.index - 1,
                "start": sub.start.total_seconds(),
                "end": sub.end.total_seconds(),
                "text": sub.content
            })
    return {"segments": segments}
```

```python
from pathlib import Path
import json

for srtfile in Path(SRT_DIR).glob("*.srt"):
    outname = srtfile.with_suffix(".json")
    segs = srt_to_json(srtfile)
    with open(outname, "w") as out:
        json.dump(segs, out, indent=2)
```
