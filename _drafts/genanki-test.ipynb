```python
import genanki

my_model = genanki.Model(
  1607392319,
  'Simple Model',
  fields=[
    {'name': 'Front'},
    {'name': 'Back'}
  ],
  templates=[
    {
      'name': 'Card 1',
      'qfmt': '{{Front}}',
      'afmt': '{{Back}}'
    }
  ])

my_deck = genanki.Deck(
  2059400110,
  'My Vocabulary Deck')

# Add cards
notes = [
    ("ubiquitous", "present everywhere"),
    ("parsimonious", "unwilling to spend money")
]

for front, back in notes:
    my_deck.add_note(genanki.Note(
        model=my_model,
        fields=[front, back]
    ))

genanki.Package(my_deck).write_to_file('vocab.apkg')
```
