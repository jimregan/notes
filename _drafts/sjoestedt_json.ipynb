```python
import json

with open("sjoestedt_phonetique.json") as inf:
    DATA = json.load(inf)
```

```python
missing_irish = []

for sect in DATA:
    for item in DATA[sect]:
        if not "irish" in item:
            missing_irish.append(item)
        else:
            if type(item["irish"]) == str and type(item["transcription"]) == str:
                print(f'{item["id"]}\t{item["irish"]}\t{item["transcription"]}')
```

```python
words = {}

missing_irish = []

for sect in DATA:
    for item in DATA[sect]:
        if not "irish" in item:
            continue
        if type(item["irish"]) == list:
            irish = item["irish"]
        else:
            irish = [item["irish"]]
        if type(item["transcription"]) == list:
            transcription = item["transcription"]
        else:
            transcription = [item["transcription"]]
        for a in irish:
            if not a in words:
                words[a] = {}
            for b in transcription:
                if not b in words[a]:
                    words[a][b] = set()
                words[a][b].add(item["id"])
```
