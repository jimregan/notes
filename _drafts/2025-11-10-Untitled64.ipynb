```python
import stanza
stanza.download('sv')
nlp = stanza.Pipeline(
    lang='sv',
    processors='tokenize,pos,lemma,depparse',
    use_gpu=True
)
```

```python
doc = nlp("Jag kan inte ge herr Allard några uppgifter om antalet barn i Umeå, och inte heller kan jag säga, hur många gymnastiklärare det finns i staden.")
conllu_text = "{:C}".format(doc)
print(conllu_text)
```
