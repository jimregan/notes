```python
def get_json_url(publication_id: str) -> str:
    return f"https://www.sverigesradio.se/playerajax/audio?id={publication_id}&type=publication&publicationid={publication_id}&quality=medium"
```

```python
def get_json(publication_id: str) -> dict:
    import requests
    import json

    json_url = get_json_url(publication_id)
    req = requests.get(json_url)
    req.raise_for_status()
    return json.loads(req.text)
```
