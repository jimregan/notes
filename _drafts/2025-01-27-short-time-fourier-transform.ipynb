```python
import matplotlib.pyplot as plt
import numpy as np
import librosa

y, sr = librosa.load(librosa.example("trumpet"), sr=None)

start_sample = 1000
end_sample = 10000
```

```python
window_start = 1500
window_end = 2012
window_length = window_end - window_start

sine_wave_samples = np.linspace(0, np.pi, window_length)
sine_wave_pulse = -0.7 + 0.07 * np.sin(sine_wave_samples)

zoomed_segment = y[start_sample:end_sample]

samples = np.arange(start_sample, end_sample)
window_samples = np.arange(window_start, window_end)

fig, axs = plt.subplots(2, 1, figsize=(12, 8))

axs[0].plot(samples, zoomed_segment, label="Original Signal", color="blue", alpha=0.7)
axs[0].fill_between(
    window_samples,
    y1=-0.7,
    y2=sine_wave_pulse,
    color="red",
    alpha=0.3,
    label="Window"
)
axs[0].set_title("Short-Time Fourier Analysis")
axs[0].set_xlabel("Samples")
axs[0].set_ylabel("Amplitude")
axs[0].set_ylim(-0.7, 0.7)
axs[0].legend()

windowed_segment = y[window_start:window_end] * np.hamming(window_length)
fft_result = np.fft.rfft(windowed_segment)
fft_magnitude_db = 10 * np.log10(np.abs(fft_result) ** 2 + 1e-10)
freqs = np.fft.rfftfreq(window_length, 1 / sr)
axs[1].plot(freqs, fft_magnitude_db, color="blue")
axs[1].set_xlim(0, 8000)
axs[1].set_ylim(-70, 30)
axs[1].set_title("FFT, window length: 512, window kind: hamming")
axs[1].set_xlabel("Frequency (Hz)")
axs[1].set_ylabel("Power Spectrum (dB)")
axs[1].grid()

plt.tight_layout()
plt.show()
```
