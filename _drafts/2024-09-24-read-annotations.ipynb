```python
espeak_to_cmudict = {}
for line in MAPPING.split("\n"):
    if line == "":
        continue
    line = line.strip()
    parts = line.split(" ")

    if len(parts) != 2:
        print(line)
        continue
    v, k = line.split(" ")
    espeak_to_cmudict[k] = v
```

```python
cmudict_to_espeak = {}
for line in MAPPING.split("\n"):
    if line == "":
        continue
    line = line.strip()
    parts = line.split(" ")

    if len(parts) != 2:
        print(line)
        continue
    k, v = line.split(" ")
    if not k in cmudict_to_espeak:
        cmudict_to_espeak[k] = v
```
