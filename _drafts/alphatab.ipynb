```python
import base64
from IPython.display import display, HTML

def display_alphatab_inline(file_path):
    # Read the binary content of the file
    with open(file_path, "rb") as file:
        binary_content = file.read()

    # Encode to base64 for embedding in a data URI
    base64_content = base64.b64encode(binary_content).decode("utf-8")
    data_uri = f"data:application/octet-stream;base64,{base64_content}"

    # JavaScript and HTML code to render AlphaTab with the embedded data URI
    html_code = f"""
    <div id="alphaTabContainer" style="width: 100%; height: 400px;"></div>

    <script type="text/javascript">
      document.addEventListener('DOMContentLoaded', function () {{
          var alphaTabContainer = document.querySelector('#alphaTabContainer');

          // Initialize AlphaTab with the container and the embedded data URI
          var api = new alphaTab.AlphaTabApi(alphaTabContainer, {{
              file: "{data_uri}",
              core: {{
                  tex: true  // Enable Tex Mode (optional)
              }}
          }});

          api.render();
      }});
    </script>
    """

    display(HTML(html_code))
```

```python
display_alphatab_inline('/content/16 june 06 3.gp5')
```
