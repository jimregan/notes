```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_url(url):
    response = requests.get(url)
    assert response.status_code == 200
    soup = BeautifulSoup(response.content, "html.parser")
    return soup
```

```python
import hashlib

def get_parse_from_selector(soup, selector):
    sub = soup.select(selector)
    assert len(sub) == 1
    text = sub[0].text
    hash = hashlib.sha256(text.encode()).hexdigest()
    retval = {
        "sha256": hash,
        "selector": selector,
        "nlp": nlp(text)
    }

    return retval
```

```python
a = get_parse_from_selector(soup, "#ajax-page-container > div.g-hidden-lg-down > div.page-text.fill-page > div > p:nth-child(3)")
```

```python
lines = "{:C}".format(a["nlp"]).split("\n")
```
