```python
import hashlib

def get_parse_from_selector(soup, selector):
    sub = soup.select(selector)
    assert len(sub) == 1
    text = sub[0].text
    hash = hashlib.sha256(text.encode()).hexdigest()
    retval = {
        "sha256": hash,
        "selector": selector,
        "nlp": nlp(text)
    }

    return retval
```
