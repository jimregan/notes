# Sine tone

Generate a sine tone (from [librosa documentation](https://librosa.org/doc/main/generated/librosa.tone.html))

```python
import librosa
tone = librosa.tone(440, sr=22050, length=22050)
```

```python
from IPython.display import Audio, display

display(Audio(tone, rate=22050))
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
S = librosa.feature.melspectrogram(y=tone)
librosa.display.specshow(librosa.power_to_db(S, ref=np.max),
                         x_axis='time', y_axis='mel', ax=ax)
```
