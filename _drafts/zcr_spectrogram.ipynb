```python
def zcr_rate(wav_in, step=240, sz=960):
    # Calculate the number of frames (steps) to compute ZCR for
    steps = (len(wav_in) - sz) // step
    
    # Initialize an array to hold ZCR values
    zrate = np.zeros(steps)
    
    # Compute ZCR for each frame
    for i in range(steps):
        start = i * step
        end = start + sz
        frame = wav_in[start:end]
        
        # Compute ZCR for the current frame
        zcr = librosa.feature.zero_crossing_rate(frame, frame_length=sz, hop_length=sz)[0, 0]
        zrate[i] = zcr
        
    return zrate
```

```python
def zcr_rate(wav_in, step=240, segment=960):
    # Calculate the sign of the signal values and shift the result to avoid zero values
    sign_wav = np.sign(wav_in + 1e-8)
    
    # Calculate the absolute difference of the sign changes, this gives 2 at zero crossings
    cross = np.abs(np.diff(sign_wav))
    
    # Clip values greater than 1 to 1 (since only sign changes are considered crossings)
    cross[cross > 1] = 1
    
    # Determine the number of segments (frames) that will be analyzed
    steps = int((np.shape(cross)[0] - segment) / step)
    
    # Initialize an array to store the zero-crossing rates for each segment
    zrate = np.zeros((steps,))
    
    # Loop over each segment to calculate the zero-crossing rate
    for i in range(steps):
        # Calculate the mean of the zero crossings in the current segment
        zrate[i] = np.mean(cross[i*step:i*step+segment])
    
    # Return the array of zero-crossing rates
    return zrate
```

```python
import numpy as np

def zero_crossing_rate(signal):
    """
    Calculate the zero-crossing rate of a signal.

    Parameters:
    signal (numpy array): Input signal array.

    Returns:
    float: Zero-crossing rate.
    """
    T = len(signal)
    # Calculate the product of each sample with the previous one
    crossings = signal[:-1] * signal[1:]
    # Count the number of times the product is negative (crossing zero)
    zcr = np.sum(crossings < 0) / (T - 1)
    return zcr

# Example usage
signal = np.array([1, -1, 1, -1, 1, -1])
zcr = zero_crossing_rate(signal)
print(f"Zero-crossing rate: {zcr}")
```

```python
import numpy as np
from numba import njit

@njit
def zcr_rate(wav_in, step=240, segment=960):
    # Calculate the sign of the signal values and shift to avoid zero values
    sign_wav = np.sign(wav_in + 1e-8)

    absdiff = np.abs(np.diff(sign_wav))

    for i in range(len(absdiff)):
        if absdiff[i] > 1:
            absdiff[i] = 1

    steps = (len(absdiff) - segment) // step

    zcrate = np.zeros(steps)

    for i in range(steps):
        start_idx = i * step
        end_idx = start_idx + segment
        zcrate[i] = np.mean(absdiff[start_idx:end_idx])

    return zcrate
```
