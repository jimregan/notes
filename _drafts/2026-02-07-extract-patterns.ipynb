```python
def strip_ref(text):
    while '&lt;ref&gt;' in text:
        ref_start = text.index('&lt;ref&gt;')
        if '&lt;/ref&gt;' in text:
            ref_end = text.index('&lt;/ref&gt;') + len('&lt;/ref&gt;')
            text = text[:ref_start] + text[ref_end:]
        else:
            text = text[:ref_start]
            break
    return text
```

```python
def strip_comments(text):
    if '&lt;!--' in text:
        comment_start = text.index('&lt;!--')
        if '--&gt;' in text:
            comment_end = text.index('--&gt;') + len('--&gt;')
            # Recursively check the remaining text
            return strip_comments(text[:comment_start] + text[comment_end:])
        else:
            return text[:comment_start]
    return text
```
