```python
def strip_ref(text):
    if '&lt;ref&gt;' in text:
        ref_start = text.index('&lt;ref&gt;')
        if '&lt;/ref&gt;' in text:
            ref_end = text.index('&lt;/ref&gt;') + len('&lt;/ref&gt;')
            return text[:ref_start] + text[ref_end:]
        else:
            return text[:ref_start]
    return text
```
