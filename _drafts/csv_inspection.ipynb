```python
import csv

def get_entries(csv_file):
    entries = []
    with open(csv_file) as f:
        reader = csv.DictReader(f)
        for row in reader:
            img_url = row["image_url"]
            img = img_url.split("/")[-1]
            prompt = row["prompt"]
            utterance = row["utterance"]
            phrase = row["phrase"]
            entries.append((phrase, utterance, img))
    return entries
```
