```python
file = "/tmp/splits-orig/swe.test"
```

```python
def retsvify_line(line):
    parts = line.strip().split(" ")
    pieces = [p.split("}") for p in parts]
    word = [p[0].replace("|", "") for p in pieces if p[0] != "_"]
    phones = [p[1].replace("|", " ") for p in pieces if p[1] != "_"]
    return "".join(word), " ".join(phones)
```

```python
outlines = []
with open(file, "r") as f, open(f"{file}.retsv", "w") as out:
    for line in f.readlines():
        word, phones = retsvify_line(line)
        out.write(f"{word}\t{phones}\n")
        
```
