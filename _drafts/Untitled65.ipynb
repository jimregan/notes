```python
PL_TEXT = []
for line in PL_WIKTIONARY_EGS.split("\n"):
    if line == "":
        continue
    parts = line.split("\t")
    rendered = parts[0]
    rendered_parts = rendered.split(" → ")
    pl = rendered_parts[1]
    pl = pl.replace("(", "").replace(")", "").strip()
    PL_TEXT.append(pl)
```

```python
egs = nlp("\n".join(PL_TEXT))
```

```python
conllu_egs = "{:C}".format(egs)
print(conllu_egs)
```
