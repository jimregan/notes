```python
import auditok
input = 'Sraith Pictiúr - An Ghaeilge Seoid luachmhar-D44-x6PTd_Q.mkv'
audio_regions = auditok.split(
    input,
    min_dur=1,     # minimum duration of a valid audio event in seconds
    max_dur=10,       # maximum duration of an event
    max_silence=0.6, # maximum duration of tolerated continuous silence within an event
    energy_threshold=25 # threshold of detection
)
for i, r in enumerate(audio_regions):

    # Regions returned by `split` have 'start' and 'end' metadata fields
    print("Region {i}: {r.meta.start:.3f}s -- {r.meta.end:.3f}s".format(i=i, r=r))
```

```python
#regs = auditok.load(input)
#regs.split_and_plot()
regs = auditok.load(input)
regs.split_and_plot(
    min_dur=1,     # minimum duration of a valid audio event in seconds
    max_dur=10,       # maximum duration of an event
    max_silence=0.6, # maximum duration of tolerated continuous silence within an event
    energy_threshold=25 # threshold of detection
)
```
