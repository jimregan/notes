```python
U = "https://www.teanglann.ie/CanU/Taoch.mp3"
```

```python
model = "jimregan/wav2vec2-fuaimeanna-phonetic"
```

```python
!wget {U}
!wget {U.replace("CanU", "CanM")}
!wget {U.replace("CanU", "CanC")}
```

```python
from transformers import pipeline
pipe = pipeline(model=model, device=0)
```

```python
pipe("Taoch.mp3")
```

```python
pipe("Taoch.mp3.1")
```

```python
pipe("Taoch.mp3.2")
```
